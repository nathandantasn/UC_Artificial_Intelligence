#Excel
O pandas pode ler e gravar arquivos do Excel. O método importa apenas dados, não fórmulas ou imagens. Ter imagens ou macros no arquivo pode fazer com que o método **read_excel** falhe

#### Excel Input

```python
pd.read_excel('Excel_Sample.xlsx',sheet_name='Sheet1')
```

#### Excel Output

```python
df.to_excel('Excel_Sample.xlsx',sheet_name='Sheet1')
```
